```python
import json
import snowflake.connector
from abc import ABCMeta, abstractmethod
import boto3

class Credentials(metaclass=ABCMeta):
    pass
    
    
class SSMPSCredentials(Credentials):
    def __init__(self, secretid: str):
        self._secretid = secretid
        self._secrets = {}
        
    def get_keys(self):
        """
        credential fetching 
        """
        _aws_sm_args = {'service_name': 'secretsmanager', 'region_name': 'us-east-1'}
        secrets_client = boto3.client(**_aws_sm_args)
        get_secret_value_response = secrets_client.get_secret_value(SecretId=self._secretid)
        return get_secret_value_response
    
    
class BaseConnector(metaclass=ABCMeta):
    @abstractmethod
    def connect(self):
        raise NotImplementedError
        

class SnowflakeConnector(BaseConnector):
    def __init__(self, credentials: Credentials):
        keys = credentials.get_keys()
        self._secrets = json.loads(keys.get('SecretString', "{}"))

    def connect(self, dbname: str, schema: str = 'DEFAULT'):
        ctx = snowflake.connector.connect(
            user=self._secrets['login_name'],
            password=self._secrets['login_password'],
            account=self._secrets['account'],
            warehouse=self._secrets['warehouse'],
            database=dbname,
            schema=schema
        )

        return ctx
```

```python
MAX_FETCHED_RESULTSET_SIZE = MAX_QUERY_RETURN_SIZE

def execute_query(query: str):
    cur = ctx.cursor()
    try:
        cur.execute(query)
        while True:
            results = cur.fetchmany(MAX_FETCHED_RESULTSET_SIZE)
            if len(results) == 0:
                break
            for row in results:
                yield row
    finally:
        cur.close()
         
def execute_insert_query(query: str):
    cur = ctx.cursor()
    try:
        cur.execute(query)
    finally:
        cur.close()

def snowflake_data_pull(query,columns):
    # DB, schema, and table names
    DB_NAME = "max_dev"
    SCHEMA = "workspace"

    # Input data from max_dev snowflake
    connector = SnowflakeConnector(SSMPSCredentials(SF_CREDS))
    ctx = connector.connect(dbname=DB_NAME, schema=SCHEMA)
    df = pd.DataFrame(execute_query(query), columns=columns)
    return df
```

```python
def mean_abs_pct_error(y_true, y_pred):
    '''Take in true and predicted and calculate MAPE'''
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred)/y_true))

year_list = [2020, 2021, 2022, 2023, 2024]
country_name = 'US'
df_holiday = make_holidays_df(year_list, country_name)

df_holiday['upper_window'] = 7
df_holiday['lower_window'] = -3
```

```python
def monthly_forecast(table, multiplier):
    table = table.fillna(0)
    table['ds_month'] = table['ds'].apply(lambda x:x.replace(day=1))
    table = table.groupby(by=['ds_month']).mean().reset_index().sort_values(by='ds_month')
    table['yhat'] = table['yhat']*multiplier
    return table[['ds_month','yhat']]

def qtr_forecast(table, multiplier, frequency):
    table = table.fillna(0)
    table['ds_'+frequency] = table['ds'].apply(lambda x:pd.Period(x, freq=frequency).start_time) #    .start_time.strftime('%y-%m'))
    table = table.groupby(by=['ds_'+frequency]).mean().reset_index().sort_values(by='ds_'+frequency)
    table['yhat'] = table['yhat']*multiplier
    return table[['ds_'+frequency,'yhat']]

def graph_monthly_forecast(table, column, tier, date_col, y_lim):
    fig, ax = plt.subplots(figsize=(7,3))
    ax.bar(table[date_col], table[column], width = 20, label= tier)
    ax.set_ylim(y_lim)
    plt.xticks(rotation=45) #, ticks = np.arange(len(table[date_col])), label = table[date_col])
    plt.title('Tier '+tier)
    rects = ax.patches
    # Make some labels.
    labels = [f"{i:.1f}" for i in table[column]] #range(len(rects))]
    for rect, label in zip(rects, labels):
        height = rect.get_height()
        ax.text(
            rect.get_x() + rect.get_width() / 2, height, label, ha="center", va="bottom"
        )    
    return

# def graph_monthly_forecast(table, column, tier, date_col):
#     fig, ax = plt.subplots(figsize=(6,3))
#     ax.bar(table[date_col], table[column], width = 20, label= tier)
#     ax.set_ylim(-.3,4)
#     plt.title('Tier '+tier)
#     plt.xticks(rotation=45)
#     plt.grid()
#     plt.tight_layout()
#     return

def graph_forecast(table, title):
    fig1 = m.plot(table)
    a = add_changepoints_to_plot(fig1.gca(), m, table)
    fig1.gca().set_ylim([-0.2, 2])
    fig1.gca().set_xlim([pd.to_datetime('2020-02-01'), pd.to_datetime('2025-01-01')])
    plt.title(title)
    fig1.set_size_inches(6, 4)
    plt.xticks(rotation=45)
    plt.tight_layout()
    return



# Save Forecasted Values
def save_forecast(df_forecast, df_historical, tier):
    fcast_t = forecast.set_index('ds')['yhat']
    actuals_t = df_historical.set_index('request_date')[tier+'_hps']
    actuals_t.index.name = 'ds'
    
    df_fcast_t = pd.DataFrame({'t'+tier+'_hps':actuals_t, 't'+tier+'_yhat':fcast_t})
    df_fcast_t['t'+tier+'_pred'] = np.where(df_fcast_t.index < pd.to_datetime(max_historical_dte),
                                           df_fcast_t['t'+tier+'_hps'], df_fcast_t['t'+tier+'_yhat'])
    return df_fcast_t

def apply_expit(table, col_list):
    for col in col_list:
        table[col] = expit(table[col])
    return table

future_window_end = pd.to_datetime('2024-12-31')
```

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import causal_lib as csl

%matplotlib inline

run_date = '2021_08_17'
max_historical_dte = pd.to_datetime('2021-08-01')
```

```python
predictions_all = csl.import_pickle('tier_predictions_'+run_date)
```

```python
predictions_all = csl.import_pickle('tier_predictions_'+run_date)

predictions_all = predictions_all.fillna(0)
predictions_all = predictions_all.replace([np.inf, -np.inf], np.nan)
predictions_all = predictions_all.reset_index(drop=False)

predictions_all['ds_month'] = predictions_all['ds'].apply(lambda x:x.replace(day=1))
predictions_all = predictions_all.groupby(by=['ds_month']).mean().reset_index().sort_values(by='ds_month')

predictions_all['t0_yhat'] = predictions_all['t0_yhat']*8.5

predictions_all['t1_yhat'] = predictions_all['t1_yhat']*8.5
predictions_all['t1_hps'] = predictions_all['t1_hps']*8.5

predictions_all['t2_yhat'] = predictions_all['t2_yhat']*8.6
predictions_all['t2_hps'] = predictions_all['t2_hps']*8.6

predictions_all['t3_yhat'] = predictions_all['t3_yhat']*8.5
predictions_all['t3_hps'] = predictions_all['t3_hps']*8.5

predictions_all['pop_yhat'] = predictions_all['pop_yhat']*10.9
predictions_all['pop_hps'] = predictions_all['pop_hps']*10.9

predictions_all = predictions_all.set_index('ds_month')

pred_qtr = predictions_all.reset_index()

for col in ['t0_yhat','t1_yhat', 't2_yhat', 't3_yhat','pop_yhat']:
    predictions_all[col] = np.where(predictions_all.index<max_historical_dte, .0, predictions_all[col])
for col in ['t1_hps', 't2_hps', 't3_hps', 'pop_hps']:
    predictions_all[col] = np.where(predictions_all.index>=max_historical_dte, .0, predictions_all[col])
```

```python
df_output = predictions_all.copy()[['t0_yhat','t1_yhat', 't1_hps', 't2_yhat', 't2_hps', 't3_yhat', 't3_hps', 'pop_yhat', 'pop_hps']]
df_output.columns = pd.MultiIndex.from_tuples([(col.split('_')[0], 'monthly_'+col.split('_')[1]) for col in df_output.columns], names=['tier', 'metric'])
df_output = df_output.stack(0).reset_index(drop=False)
df_output = df_output.rename(columns = {'ds_month':'period', 'monthly_hps':'psi_tier_actual_monthly', 'monthly_yhat':'psi_tier_prediction_monthly' })
```

```python
# Adjust Quarter Forecast
pred_qtr = csl.import_pickle('tier_predictions_'+run_date)  
 
pred_qtr = pred_qtr.fillna(0)
pred_qtr = pred_qtr.replace([np.inf, -np.inf], np.nan)
pred_qtr = pred_qtr.reset_index(drop=False)
pred_qtr['ds_qtr'] = pred_qtr['ds'].apply(lambda x: str(x.year)+' Q'+str(x.quarter))

pred_qtr = pred_qtr.groupby(by=['ds_qtr']).agg({'ds':'min', 't0_yhat':'mean','t1_yhat':'mean', 't1_hps':'mean', 't2_yhat':'mean', 't2_hps':'mean',
                                    't3_yhat':'mean', 't3_hps':'mean', 'pop_yhat':'mean', 'pop_hps':'mean'})

pred_qtr['t0_yhat'] = pred_qtr['t0_yhat']*17.6

pred_qtr['t1_yhat'] = pred_qtr['t1_yhat']*17.6
pred_qtr['t1_hps'] = pred_qtr['t1_hps']*17.6

pred_qtr['t2_yhat'] = pred_qtr['t2_yhat']*18.5
pred_qtr['t2_hps'] = pred_qtr['t2_hps']*18.5

pred_qtr['t3_yhat'] = pred_qtr['t3_yhat']*17.8
pred_qtr['t3_hps'] = pred_qtr['t3_hps']*17.8

pred_qtr['pop_yhat'] = pred_qtr['pop_yhat']*23.2
pred_qtr['pop_hps'] = pred_qtr['pop_hps']*23.2

for col in ['t1_yhat', 't2_yhat', 't3_yhat','pop_yhat']:
    pred_qtr[col] = np.where(pred_qtr['ds']<pd.Period(max_historical_dte, freq='Q').start_time, .0, pred_qtr[col])
for col in ['t1_hps', 't2_hps', 't3_hps', 'pop_hps']:
    pred_qtr[col] = np.where(pred_qtr['ds']>=pd.Period(max_historical_dte, freq='Q').start_time, .0, pred_qtr[col])
```

```python
pred_qtr = pred_qtr.reset_index(drop=False)

df_qtr_output = pred_qtr.copy()[['ds','t0_yhat','t1_yhat', 't1_hps', 't2_yhat', 't2_hps', 't3_yhat', 't3_hps', 'pop_yhat', 'pop_hps']]
df_qtr_output = df_qtr_output.set_index('ds')
df_qtr_output.columns =  pd.MultiIndex.from_tuples([(col.split('_')[0], 'quarterly_'+col.split('_')[1]) for col in df_qtr_output.columns], names=['tier', 'metric'])
df_qtr_output = df_qtr_output.stack(0).reset_index(drop=False)
df_qtr_output = df_qtr_output.rename(columns = {'ds':'period', 'quarterly_hps':'psi_tier_actual_quarterly', 'quarterly_yhat':'psi_tier_prediction_quarterly' })
```

```python
df_qtr_output = df_qtr_output.set_index(['period','tier'])

df_output = df_output.set_index(['period','tier'])
df_output = df_output.join(df_qtr_output,  how='left').reset_index(drop=False)
df_output = df_output[df_output['period']<='2024-12-31']
df_output['schedule_label'] = 'alpha'
# df_output.to_csv('psi_predictions_'+run_date+'all'.csv', index=False)
```

```python
col_list = ['psi_tier_actual_monthly','psi_tier_prediction_monthly', 
            'psi_tier_actual_quarterly','psi_tier_prediction_quarterly']
for col in col_list:
    df_output[col] = np.where(df_output[col]==0, None, df_output[col])
```

```python
# Format output table
def graph_predictions(table, hps_list, prediction_list, tier_label, colors, title_name, bar_width, x_lim, date_col):
    table['t0_hps'] = np.nan
    fig, ax = plt.subplots(figsize=(10,6))
    fig.set_facecolor('white')
    bottom_hps = table[hps_list[0]]
    bottom_prediction = table[prediction_list[0]]
    for num in np.arange(len(hps_list)):
        ax.bar(table.index, table[hps_list[num]], label='Tier '+tier_label[num], 
               width=bar_width, edgecolor=colors[num], color=colors[num],
               bottom=bottom_hps-table[hps_list[0]])
        ax.bar(table.index, table[prediction_list[num]], label='Tier '+tier_label[num]+' Prediction', 
               width=bar_width, edgecolor='darkblue', color=colors[num], alpha=.3,
               bottom=bottom_prediction-table[prediction_list[0]])
        
        bottom_hps=bottom_hps.add(table[hps_list[num]], fill_value=0)
        bottom_prediction=bottom_prediction.add(table[prediction_list[num]], fill_value=0)
#     plt.xticks(rotation = 45, ticks = np.arange(len(pred_qtr['ds_qtr'])), labels=pred_qtr['ds_qtr'])
    plt.xticks(rotation=45, ticks=np.arange(len(table[date_col])), labels=table[date_col])
    plt.title(title_name)
    plt.ylim(0,20)
#     plt.xlim(x_lim)
    plt.legend(loc='upper left')
    plt.tight_layout()   
    
# graph_predictions(predictions_all, hps_list, prediction_list, tier_label, colors, 'Monthly Hours per Viewer - Budget 1')
```
