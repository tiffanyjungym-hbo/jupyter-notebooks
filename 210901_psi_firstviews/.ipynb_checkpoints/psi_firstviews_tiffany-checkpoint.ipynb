## Feature Engineering

```python
def get_quarter(df, timestamp='EFFECTIVE_START_DATE'):
    return pd.to_datetime(df[timestamp]).apply(lambda x: x.quarter)

prefix_dicts = {
    'CONTENT_SOURCE': 'brand',
    'CONTENT_CATEGORY': 'content_cat',
    'quarter': 'Q',
    'PROGRAM_TYPE': 'program',
    'CATEGORY': 'cat',
    'TIER': 'tier',
    'SEASON_NUMBER': 'sn',
    'PREMIERING_HOURS_RUNTIME': 'runtime'
}

def get_one_hot_encode(df, column, prefix):
    text = df[column]
    text = text.apply(lambda x: x.lower().replace(' ', '_'))
    
    return pd.get_dummies(text, prefix=prefix)
```

```python
df_feat = df_agg.copy()
```

```python
def feature_engineering(df):
    df_ = df.copy()
    
    df_['EFFECTIVE_START_DATE'] = pd.to_datetime(df_['EFFECTIVE_START_DATE'], format='%Y-%m-%d')
    df_['quarter'] = get_quarter(df_, timestamp='EFFECTIVE_START_DATE')

    OHE_FINAL_COLUMNS = []
    for ohe_column in ['CONTENT_SOURCE', 'quarter', 'CATEGORY', 'TIER', 'CONTENT_CATEGORY']:

        df_[ohe_column] = df_[ohe_column].astype(str)
        ohes = get_one_hot_encode(df_, column=ohe_column, prefix=prefix_dicts[ohe_column])
        df_ = pd.concat([df_, ohes], axis=1)

        OHE_FINAL_COLUMNS.extend(ohes.columns.tolist())
    
    return df_, OHE_FINAL_COLUMNS
```
