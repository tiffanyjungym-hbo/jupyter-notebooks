```python
import os
import sys
path=!pwd
sys.path.append(os.path.join(path[0], '..'))
import numpy as np
import math
import pandas as pd
import re
from datetime import date, datetime, timedelta
import json
from abc import ABCMeta, abstractmethod
import boto3
import snowflake.connector
from io import StringIO
import logging 

logger = logging.getLogger()
logging.basicConfig(level=logging.INFO)
logger.info(f'Starting Notebook')

from utils import *

class Utils():
    @staticmethod
    def to_csv_s3(content, bucket, key_path, filename):
        client = boto3.client('s3')
        key = os.path.join(key_path, filename)
        csv_buffer = StringIO()
        content.to_csv(csv_buffer)
        client.put_object(Bucket=bucket, Key=key, Body=csv_buffer.getvalue())
        logger.info(f'Saved to {bucket}/{key}')
    
    @staticmethod
    def to_pkl_s3(content, bucket, key_path, filename):
        client = boto3.client('s3')
        key = os.path.join(key_path, filename)
        obj = pickle.dumps(content)
        client.put_object(Bucket=bucket, Key=key, Body=obj)
        logger.info(f'Saved model to {os.path.join(bucket, key)}')
        logger.info(f'Saved to {bucket}/{key}')

    @staticmethod
    def read_csv_s3(bucket, key_path,filename):
        client = boto3.client('s3')
        key = os.path.join(key_path, filename)
        obj = client.get_object(Bucket=bucket, Key=key)
        df = pd.read_csv(obj['Body'], na_values="\\N")
        logger.info(f'Read from {bucket}/{key}')
        return df
        
    @staticmethod
    def read_pkl_s3(bucket, key_path,filename):
        client = boto3.client('s3')
        key = os.path.join(key_path, filename)
        obj = client.get_object(Bucket=bucket, Key=key)
        body = obj['Body'].read()
        model = pickle.loads(body)
        logger.info(f'Read from {bucket}/{key}')
        return model    

class Credentials(metaclass=ABCMeta):
    pass
    
    
class SSMPSCredentials(Credentials):
    def __init__(self, secretid: str):
        self._secretid = secretid
        self._secrets = {}
        
    def get_keys(self):
        """
        credential fetching 
        """
        _aws_sm_args = {'service_name': 'secretsmanager', 'region_name': 'us-east-1'}
        secrets_client = boto3.client(**_aws_sm_args)
        get_secret_value_response = secrets_client.get_secret_value(SecretId=self._secretid)
        return get_secret_value_response
    
    
class BaseConnector(metaclass=ABCMeta):
    @abstractmethod
    def connect(self):
        raise NotImplementedError

        
class SnowflakeConnector(BaseConnector):
    def __init__(self, credentials: Credentials):
        keys = credentials.get_keys()
        self._secrets = json.loads(keys.get('SecretString', "{}"))

    def connect(self, dbname: str, schema: str = 'DEFAULT'):
        ctx = snowflake.connector.connect(
            user=self._secrets['login_name'],
            password=self._secrets['login_password'],
            account=self._secrets['account'],
            warehouse=self._secrets['warehouse'],
            database=dbname,
            schema=schema
        )
        return ctx

    
def run_query(querystr, ctx):
    cursor_list = ctx.execute_string(
        querystr
        )
    df = pd.DataFrame.from_records(cursor_list[-1].fetchall(), columns=[x[0] for x in cursor_list[-1].description])
    df.columns= df.columns.str.lower()
    return df

import plotly.express as px

def get_simple_plot(df_plt, x, y, grpby, text, title=''):
    if title=='':
        title = f'{y} vs {x}'
    df_plt[grpby] = df_plt[grpby].astype(str)
    fig = px.line(df_plt,
                  x=x, 
                  y=y, 
                  title=title,
                  color=grpby, 
                  hover_data=[text],
                  width=800, height=400)
    fig.show()
    return 


input_bucket="hbo-ingest-datascience-content-dev"
output_bucket="hbo-outbound-datascience-content-dev"
key_path = 'cost_allocation/dev'

## Credentials
SF_CREDS = 'datascience-max-dev-sagemaker-notebooks'

## Snowflake connection 
conn=SnowflakeConnector(SSMPSCredentials(SF_CREDS))
ctx=conn.connect("MAX_PROD","DATASCIENCE_STAGE")
cur = ctx.cursor()

```

```python
# churn_hv_pillar_query='''
# with air_table as (
#   select
#     viewable_id
#   , pillar_type
#   , pillar_genre
#   , row_number() over(partition by viewable_id order by release desc, earliest_available_date desc) AS row_number
#   from max_prod.catalog.airtable_content_strategy
# )
# , user_hourview as (
#   select 
#       ds.hbo_uuid
#       , case when ds.subscription_cancel_ts>='2021-12-01' then 0 else 1 end as is_cancel --canceled next month  
#       , ac.pillar_genre as genre
#       , ac.pillar_type as pillar_type
#       , sum(hb.stream_elapsed_play_seconds/3600) as hours_viewed
#   from max_prod.bi_analytics.dimension_subscription ds
#   left join max_prod.viewership.max_user_stream_heartbeat hb
#       on hb.hbo_uuid = ds.hbo_uuid
#   left join air_table ac
#        on hb.viewable_id = ac.viewable_id  
#   where ac.row_number=1 
#   and ds.subscription_start_ts <='2021-09-01'
#   and ds.subscription_cancel_ts >='2021-11-01'
#   and ds.original_territory='HBO MAX DOMESTIC'
#   and ds.is_retail = 'TRUE'
#   and hb.request_date >='2021-10-01'
#   and hb.request_date <='2021-11-01'
#   and hb.viewable_id IS NOT NULL 
#   and hb.stream_elapsed_play_seconds >= 120
#   and hb.video_type = 'main'
#   and hb.channel = 'HBO MAX SUBSCRIPTION'
#   and hb.country_iso_code in ('US')
#   group by 1,2,3,4
# )
# select 
#     uh.genre
#     , uh.pillar_type
#     , round(uh.hours_viewed, 0) as hours_viewed
#     , sum(uh.is_cancel) as total_cancel
#     , count(distinct(uh.hbo_uuid)) as total_viewer
# from user_hourview uh
# group by 1,2,3
# '''

churn_hv_query='''
with user_hourview as (
  select 
      ds.hbo_uuid
      , case when ds.subscription_cancel_ts>='2021-12-01' then 0 else 1 end as is_cancel --canceled next month  
      , rad.primary_genre_desc as genre
      , rad.content_category as content_category
      , sum(hb.stream_elapsed_play_seconds/3600) as hours_viewed
  from max_prod.bi_analytics.dimension_subscription ds
  left join max_prod.viewership.max_user_stream_heartbeat hb
      on hb.hbo_uuid = ds.hbo_uuid
  left join max_prod.catalog.reporting_asset_dim rad
      on hb.viewable_id = rad.viewable_id
  where ds.subscription_start_ts <='2021-09-01'
  and ds.subscription_cancel_ts >='2021-11-01'
  and ds.original_territory='HBO MAX DOMESTIC'
  and ds.is_retail = 'TRUE'
  and hb.request_date >='2021-10-01'
  and hb.request_date <='2021-11-01'
  and hb.viewable_id IS NOT NULL 
  and hb.stream_elapsed_play_seconds >= 120
  and hb.video_type = 'main'
  and hb.channel = 'HBO MAX SUBSCRIPTION'
  and hb.country_iso_code in ('US')
  group by 1,2,3,4
)
select 
    uh.genre
    , uh.content_category
    , round(uh.hours_viewed, 0) as hours_viewed
    , sum(uh.is_cancel) as total_cancel
    , count(distinct(uh.hbo_uuid)) as total_viewer
from user_hourview uh
group by 1,2,3
'''

df_hv =  run_query(churn_hv_query, ctx)
Utils.to_csv_s3(df_hv, input_bucket, key_path, 'churn_hv_20211001.csv')
```

```python
## Example case for drama series and kids/family movie 
from pyomo.environ import *
import numpy as np

def get_b(x, e, c):
    b = x * c / e
    return b

## content efficiency, churn slope curve 
e_dic = {'drama_series': 0.05, 'kids/family_movie': 0.1}  ## efficiency 
c_dic = {'drama_series': 1, 'kids/family_movie': 10} # $M/hour, cost 
a_coeff_dic = {'drama_series': -0.0076, 'kids/family_movie': -0.0011} ## from curve fitting 
b_coeff_dic = {'drama_series': -0.507, 'kids/family_movie': -0.071} ## from curve fitting 
# a_coeff_dic = {'drama_series': -0.005, 'kids/family_movie': -0.005}
# b_coeff_dic = {'drama_series': -0.507, 'kids/family_movie': -0.507}
b0 = 300
x = list(np.linspace(0, 10, num=11))

i = 'drama_series'
bval1 = [get_b(j, e_dic[i], c_dic[i]) for j in x]
churn_slope1 = [a_coeff_dic[i]*np.exp(b_coeff_dic[i]*j) for j in x]

i = 'kids/family_movie'
bval2 = [get_b(j, e_dic[i], c_dic[i]) for j in x]
churn_slope2 = [a_coeff_dic[i]*np.exp(b_coeff_dic[i]*j) for j in x]


## define model
model = ConcreteModel()

# Var
model.b1 = Var(domain=NonNegativeReals, bounds=(0,500), initialize = b0)
model.b2 = Var(domain=NonNegativeReals, bounds=(0,500), initialize = b0)
model.y1 = Var(bounds=(-1,0))
model.y2 = Var(bounds=(-1,0))

# Constraints 
model.con1 = Piecewise(model.y1, model.b1, 
                                pw_pts=bval1,  # domain breakpoints for the piecewise linear function
                                f_rule=churn_slope1, #ange value corresponding to each piecewise domain point
                                pw_constr_type='EQ',  #bound type of the piecewise function. EQ=y variable is equal to the piecewise function
                                pw_repn='DCC') # domain breakpoints for the piecewise linear function
model.con2 = Piecewise(model.y2, model.b2, 
                                pw_pts=bval2,  # domain breakpoints for the piecewise linear function
                                f_rule=churn_slope2, #ange value corresponding to each piecewise domain point
                                pw_constr_type='EQ',  #bound type of the piecewise function. EQ=y variable is equal to the piecewise function
                                pw_repn='DCC') # domain breakpoints for the piecewise linear function

model.con3 = Constraint(expr= model.b1 + model.b2 <= 1000)
model.con4 = Constraint(expr= model.b1 + model.b2 >= 50)
model.con5 = Constraint(expr= model.y1 <= -0.000001)
model.con6 = Constraint(expr= model.y2 <= -0.000001)
model.con7 = Constraint(expr= model.y1 >= -0.01)
model.con8 = Constraint(expr= model.y2 >= -0.01)
model.con9 = Constraint(expr= model.b1 >= 10)
model.con10 = Constraint(expr= model.b2 >= 10)

# objective
model.obj = Objective(expr=model.y1 + model.y2, sense=minimize)
# results = SolverFactory('glpk').solve(model)
# print(model.obj())
solver = SolverFactory('glpk')
results =solver.solve(model, tee=True)
print('obj: ', model.obj())
print('b1: ', model.b1())
print('b2: ', model.b2())
print('y1: ', model.y1())
print('y2: ', model.y2())
print('drama series hours: ', model.b1()/c_dic['drama_series'])
print('kidsfamily movie hours: ', model.b2()/c_dic['kids/family_movie'])
```
