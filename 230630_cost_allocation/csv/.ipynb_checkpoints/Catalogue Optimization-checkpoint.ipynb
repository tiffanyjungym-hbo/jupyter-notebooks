```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('churn_curves.csv')
df = df.set_index(['PILLAR_GENRE'])
df = df[df['HOURS_WATCHED'] <= 100]
df = df[df['CHURN_RATE'] < 1]
df = df.sort_values('HOURS_WATCHED', ascending=True)
```

```python
df
```

```python
df_hrs = pd.read_csv('hours_watched.csv')
df_hrs = df_hrs.set_index('PILLAR_GENRE')
df_hrs['users_per_title'] = df_hrs['NUMBER_USERS'] / df_hrs['TOTAL_TITLES']
```

```python
df_hrs.sort_values('AVG_HOURS')
```

```python
df
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
axt = ax.twinx()

pillars = ['Drama', 'Documentary', 'Comedy', 'Action/Adventure']

for pillar in pillars:
    p = ax.errorbar(
        df[df['NUM_SUBS'] > 500].loc[pillar, 'HOURS_WATCHED'],
        df[df['NUM_SUBS'] > 500].loc[pillar, 'CHURN_RATE'],
        label=pillar
    )

    col = p[0].get_color()
    ax.axvline(df_hrs.loc[pillar, 'AVG_HOURS'], ls='--', color=col)
    

    axt.errorbar(
        df.loc[pillar, 'HOURS_WATCHED'],
        df.loc[pillar, 'NUM_SUBS'],
        color=col,
        ls=':'
    )

axt.set_yscale('log')


leg = ax.legend(loc=0)
```
